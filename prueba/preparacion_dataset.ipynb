```python
import pandas as pd
import numpy as np

# 1. Cargar el dataset original (REPS)
df = pd.read_csv('Registro_Especial_de_Prestadores_y_Sedes_de_Servicios_de_Salud_20260412.csv')

# 2. Filtrar por Independientes en Cali
df_cali = df[(df['ClasePrestadorDesc'] == 'Profesional Independiente') & 
             (df['MunicipioSedeDesc'].str.upper() == 'CALI')].copy()

# 3. ASIGNACIÓN DE ESPECIALIDADES (Laboratorio)
especialidades_lista = [
    'Cardiología', 'Dermatología', 'Pediatría', 'Ginecología', 
    'Ortopedia', 'Oftalmología', 'Psicología', 'Odontología',
    'Urología', 'Neurología', 'Fisioterapia', 'Nutrición'
]
np.random.seed(42)
df_cali['Especialidad_Lab'] = np.random.choice(especialidades_lista, size=len(df_cali))

# 4. GENERACIÓN DE COORDENADAS (Laboratorio)
# Rango urbano aproximado de Cali
df_cali['latitud'] = np.random.uniform(3.33, 3.48, size=len(df_cali))
df_cali['longitud'] = np.random.uniform(-76.55, -76.48, size=len(df_cali))

# 5. CREACIÓN DEL PERFIL DE BÚSQUEDA (Metadata)
df_cali['Perfil_Busqueda'] = (
    df_cali['NombrePrestador'] + " " + 
    df_cali['Especialidad_Lab'] + " " + 
    df_cali['NombreSede']
).str.lower()

# 6. GUARDAR DATASET DEFINITIVO
columnas_finales = [
    'NombrePrestador', 'NombreSede', 'Especialidad_Lab', 
    'latitud', 'longitud', 'Perfil_Busqueda', 'DireccionSede'
]
df_cali[columnas_finales].to_csv('dataset_medicos_lab.csv', index=False)

print(f"✅ Dataset exportado con {len(df_cali)} registros y todas las columnas necesarias.")
```

```python
# Ver el resultado
print(f"Asignación completada para {len(df_cali)} registros.")
print("\nConteo por especialidad asignada:")
print(df_cali['Especialidad_Lab'].value_counts())
```
