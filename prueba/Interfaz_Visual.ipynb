```python
import pandas as pd
import folium
import ipywidgets as widgets
from IPython.display import display, clear_output
from motor_ia import recomendar_medico_completo # O la función local

# Carga limpia: si falla aquí, es porque el archivo no existe
df_cali = pd.read_csv('dataset_medicos_lab.csv')

def on_click_buscar(b):
    with out:
        clear_output()
        u_lat, u_lon = 3.376, -76.533 # Unicentro
        
        # El motor ahora recibirá un dataframe que garantizamos tiene latitud y Especialidad_Lab
        resultados, esp_detectada = recomendar_medico_completo(txt_sintomas.value, u_lat, u_lon, df_cali)
        
        print(f"🎯 Especialidad sugerida: {esp_detectada}")
        display(resultados[['NombrePrestador', 'Especialidad_Lab', 'distancia_km']])
        
        m = folium.Map(location=[u_lat, u_lon], zoom_start=13)
        for _, r in resultados.iterrows():
            folium.Marker([r['latitud'], r['longitud']], popup=r['NombrePrestador']).add_to(m)
        display(m)

# (Resto del código de widgets...)

# 2. Widgets de la Interfaz
txt_sintomas = widgets.Textarea(
    placeholder='Describa aquí sus síntomas...',
    description='Síntomas:', layout={'width': '100%', 'height': '100px'}
)
btn_buscar = widgets.Button(description='Buscar Especialista', button_style='success')
out = widgets.Output()

def on_click_buscar(b):
    with out:
        clear_output()
        # Ubicación fija para el lab (Cali - Unicentro)
        u_lat, u_lon = 3.376, -76.533
        
        # Llamar a la función importada
        resultados, esp_detectada = recomendar_medico_completo(txt_sintomas.value, u_lat, u_lon, df_cali)
        
        print(f"🎯 Especialidad sugerida: {esp_detectada}")
        display(resultados[['NombrePrestador', 'Especialidad_Lab', 'distancia_km']])
        
        # Crear Mapa
        m = folium.Map(location=[u_lat, u_lon], zoom_start=13)
        folium.Marker([u_lat, u_lon], popup="Tu ubicación", icon=folium.Icon(color='red')).add_to(m)
        
        for _, row in resultados.iterrows():
            folium.Marker(
                [row['latitud'], row['longitud']],
                popup=f"{row['NombrePrestador']}\n{row['distancia_km']:.2f} km",
                icon=folium.Icon(color='blue', icon='plus')
            ).add_to(m)
        display(m)

btn_buscar.on_click(on_click_buscar)
display(txt_sintomas, btn_buscar, out)
```
